# MAQM Section 3.3 — Spin-1 Angular Momentum

This notebook accompanies **Section 3.3** of *A Modern Approach to Quantum Mechanics* (Townsend).

For a spin-1 particle all operators are $3\times3$ matrices with $\hbar = 1$.
The basis is ordered $|1,{+1}\rangle,\;|1,0\rangle,\;|1,{-1}\rangle$.

## 1. The $J^2$ Operator

The **Casimir operator** $J^2 = J_x^2 + J_y^2 + J_z^2$ measures total squared angular momentum.
For a spin-$j$ particle (with $\hbar = 1$):

$$J^2 = j(j+1)\,\mathbf{I}$$

For spin-1 ($j = 1$):

$$J^2 = 1(1+1)\,\mathbf{I}_3 = 2\,\mathbf{I}_3$$

## 2. The Matrices $J_x$, $J_y$, $J_z$

The spin-1 angular momentum matrices in the $J_z$ eigenbasis are:

$$J_x = \frac{1}{\sqrt{2}}\begin{pmatrix}0&1&0\\1&0&1\\0&1&0\end{pmatrix}, \qquad
J_y = \frac{1}{\sqrt{2}}\begin{pmatrix}0&-i&0\\i&0&-i\\0&i&0\end{pmatrix}, \qquad
J_z = \begin{pmatrix}1&0&0\\0&0&0\\0&0&-1\end{pmatrix}$$

These satisfy the SU(2) commutation relations $[J_i, J_j] = i\,\varepsilon_{ijk}\,J_k$.

## 3. The Eigenstates $|1,m\rangle$

The three $J_z$ eigenstates for $j = 1$ are the standard basis vectors:

$$|1,{+1}\rangle = \begin{pmatrix}1\\0\\0\end{pmatrix}, \qquad
|1,0\rangle = \begin{pmatrix}0\\1\\0\end{pmatrix}, \qquad
|1,{-1}\rangle = \begin{pmatrix}0\\0\\1\end{pmatrix}$$

## 4. Commutators $[J^2, J_i] = 0$

$J^2$ is a **Casimir operator**: it commutes with every component of $\mathbf{J}$,

$$[J^2,\, J_x] = 0, \qquad [J^2,\, J_y] = 0, \qquad [J^2,\, J_z] = 0$$

This means $J^2$ and any single component $J_i$ share a common eigenbasis — exactly the states $|j,m\rangle$ we are using.

## 5. Eigenvalue Equations for $J^2$ and $J_z$

The states $|1,m\rangle$ simultaneously satisfy:

$$J^2\,|1,m\rangle = j(j+1)\,|1,m\rangle = 2\,|1,m\rangle \quad \text{for all } m$$

$$J_z\,|1,m\rangle = m\,|1,m\rangle$$

so explicitly:

| Eigenstate | $J^2\,|1,m\rangle$ | $J_z\,|1,m\rangle$ |
|---|---|---|
| $|1,{+1}\rangle$ | $2\,|1,{+1}\rangle$ | $+1\cdot|1,{+1}\rangle$ |
| $|1,0\rangle$ | $2\,|1,0\rangle$ | $0\cdot|1,0\rangle$ |
| $|1,{-1}\rangle$ | $2\,|1,{-1}\rangle$ | $-1\cdot|1,{-1}\rangle$ |

## 6. Ladder Operators $J_+$ and $J_-$

The **raising** and **lowering** operators are defined as:

$$J_+ = J_x + i\,J_y \qquad J_- = J_x - i\,J_y$$

They are precomputed in the toolkit as `jplus` and `jminus`, but it is instructive to build them yourself.

## 7. Action of $J_+$ on the Eigenstates

The raising operator increases $m$ by one according to:

$$J_+\,|j,m\rangle = \sqrt{j(j+1) - m(m+1)}\;|j,m+1\rangle$$

For $j = 1$:

| Input state | $J_+\,|1,m\rangle$ | Expected result |
|---|---|---|
| $|1,{+1}\rangle$ | $\sqrt{1(2)-(+1)(+2)}\,|1,2\rangle$ | $\mathbf{0}$ — already at top |
| $|1,0\rangle$ | $\sqrt{1(2)-(0)(+1)}\,|1,{+1}\rangle$ | $\sqrt{2}\,|1,{+1}\rangle$ |
| $|1,{-1}\rangle$ | $\sqrt{1(2)-(-1)(0)}\,|1,0\rangle$ | $\sqrt{2}\,|1,0\rangle$ |

## 8. Action of $J_-$ on the Eigenstates

The lowering operator decreases $m$ by one according to:

$$J_-\,|j,m\rangle = \sqrt{j(j+1) - m(m-1)}\;|j,m-1\rangle$$

For $j = 1$:

| Input state | $J_-\,|1,m\rangle$ | Expected result |
|---|---|---|
| $|1,{+1}\rangle$ | $\sqrt{1(2)-(+1)(0)}\,|1,0\rangle$ | $\sqrt{2}\,|1,0\rangle$ |
| $|1,0\rangle$ | $\sqrt{1(2)-(0)(-1)}\,|1,{-1}\rangle$ | $\sqrt{2}\,|1,{-1}\rangle$ |
| $|1,{-1}\rangle$ | $\sqrt{1(2)-(-1)(-2)}\,|1,-2\rangle$ | $\mathbf{0}$ — already at bottom |
