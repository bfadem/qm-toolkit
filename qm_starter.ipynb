## Quick Reference

Every object below is available in two forms:
- **NumPy** — numerical, e.g. `sigma_x`
- **SymPy** — symbolic (exact), e.g. `sigma_x_sym`

---

### Spin-1/2 Objects

| Object | Description |
|---|---|
| `sigma_x`, `sigma_y`, `sigma_z` | Pauli matrices |
| `sigma_vec` | List `[sigma_x, sigma_y, sigma_z]` |
| `sigma_plus`, `sigma_minus` | Ladder operators σ± = (1/2)(σ_x ± iσ_y) |
| `I2` | 2×2 identity matrix |
| `zplus`, `zminus` | σ_z eigenstates \|+z⟩, \|−z⟩ |
| `xplus`, `xminus` | σ_x eigenstates \|+x⟩, \|−x⟩ |
| `yplus`, `yminus` | σ_y eigenstates \|+y⟩, \|−y⟩ |

### Spin-1 Objects (ℏ = 1)

| Object | Description |
|---|---|
| `jx`, `jy`, `jz` | Spin-1 angular momentum matrices |
| `j_vec` | List `[jx, jy, jz]` |
| `jplus`, `jminus` | Ladder operators j± = jx ± i·jy |
| `I3` | 3×3 identity matrix |
| `jp1`, `j0`, `jm1` | jz eigenstates for m = +1, 0, −1 |

### Photon Polarization

| Object | Description |
|---|---|
| `pol_x`, `pol_y` | Linear polarization states \|x⟩, \|y⟩ |
| `pol_r`, `pol_l` | Circular polarization states \|r⟩, \|l⟩ |

### Operations Reference

All examples below work for both NumPy and SymPy objects.

#### Apply an operator to a ket  A|ψ⟩
```python
sigma_z @ zplus          # NumPy:  σ_z |+z⟩
sigma_z_sym * zplus_sym  # SymPy:  σ_z |+z⟩
```

#### Bra acting on a ket  ⟨φ|ψ⟩  (inner product)
```python
dagger(zplus) @ zminus          # NumPy:  ⟨+z|−z⟩
zplus_sym.H * zminus_sym        # SymPy:  ⟨+z|−z⟩
```

#### Bra acting on operator then ket  ⟨φ|A|ψ⟩
```python
dagger(zplus) @ sigma_x @ zminus          # NumPy
zplus_sym.H * sigma_x_sym * zminus_sym    # SymPy
```

#### Multiply two operators  AB
```python
sigma_x @ sigma_y          # NumPy
sigma_x_sym * sigma_y_sym  # SymPy
```

#### Outer product  |ψ⟩⟨φ|
```python
zplus @ dagger(zminus)      # NumPy:  |+z⟩⟨−z|
zplus_sym * zminus_sym.H    # SymPy:  |+z⟩⟨−z|
```

#### Commutator  [A, B] = AB − BA
```python
comm(sigma_x, sigma_y)          # NumPy  → 2i·σ_z
comm(sigma_x_sym, sigma_y_sym)  # SymPy  → 2I·σ_z
```

#### Expectation value  ⟨ψ|A|ψ⟩
```python
expect(sigma_z, zplus)          # NumPy  → 1.0
expect(sigma_z_sym, zplus_sym)  # SymPy  → 1
```

#### Norm  √⟨ψ|ψ⟩
```python
norm(zplus)       # NumPy  → 1.0
norm(xplus_sym)   # SymPy  → 1
```

#### Normalize a state  ψ / ‖ψ‖
```python
normalize(np.array([[3],[4]], dtype=complex))  # NumPy  → [[0.6],[0.8]]
normalize(sp.Matrix([[1],[1]]))                 # SymPy  → (1/√2)[[1],[1]]
```

#### Transition probability  |⟨φ|ψ⟩|²
```python
prob(zplus, xplus)      # NumPy  → 0.5
prob(zplus_sym, xplus_sym)  # SymPy  → 1/2
```

#### Conjugate transpose (dagger)  A†
```python
dagger(sigma_y)       # NumPy
sigma_y_sym.H         # SymPy
```

#### Scalar multiplication
```python
2j * sigma_z                    # NumPy:  2i·σ_z
2 * sp.I * sigma_z_sym          # SymPy:  2i·σ_z
```

#### Print a SymPy result as LaTeX
```python
display(sigma_x_sym)
```

#### List all available objects and operations
```python
list_objects()     # all matrices and state vectors
list_operations()  # all helper functions
```
